## 칼라 영상으로부터 r channel, g channel, b channel 별로 구분한 후 각 채널별 256 단계로 히스트로그램 및 normalized histogram을 구하기 위한 함수를 histogram_256, normal_histogram_256의 이름을 가지도록 코드를 작성한 후 각 채널별 히스토그램을 가시화 하시오. 

```
r_img = cx_train[40][:, :, 0]
g_img = cx_train[40][:, :, 1]
b_img = cx_train[40][:, :, 2]

M = 32
N = 32
L = 256

def histogram_256(channel, color):

  #배열 초기화
  array=[0 for i in range(L)]
  
  #히스토그램 칸을 1만큼 증가
  for i in range(0, M):
    for j in range(0, N):
      array[channel[i,j]]+=1
  
  #히스토그램으로 나타내기
  x = [x for x in range(256)]
  plt.plot(x, array, color=color)
  plt.xticks(range(0, 256, 50))
  plt.title("Histogram")
  plt.show()

def normal_histogram_256(channel, color):
  
  #배열 초기화
  array=[0 for i in range(L)]
  
  #히스토그램 칸을 1만큼 증가
  for i in range(0, M):
    for j in range(0, N):
      array[channel[i,j]]+=1

  #정규화
  for i in range(L):
    array[i] = array[i]/(M*N)
  
  #히스토그램으로 나타내기
  x = [x for x in range(256)]
  plt.plot(x, array, color=color)
  plt.xticks(range(0, 256, 50))
  plt.title("Normalized Histogram")
  plt.show()

histogram_256(r_img, "red")
histogram_256(g_img, "green")
histogram_256(b_img, "blue")

normal_histogram_256(r_img, "red")
normal_histogram_256(g_img, "green")
normal_histogram_256(b_img, "blue")
```

## 각 r channel, g channel, b channel별 16 단계로 단계로 히스트로그램 및 normalized histogram을 구하기 위한 함수를 histogram_256, normal_histogram_256의 이름을 가지도록 코드를 작성한 후 각 채널별 히스토그램을 가시화 하시오.

```
r_img = cx_train[40][:, :, 0]
g_img = cx_train[40][:, :, 1]
b_img = cx_train[40][:, :, 2]

M = 32
N = 32
L = 16

def histogram_16(channel, color):

  #배열 초기화
  array=[0 for i in range(L)]
  
  #히스토그램 칸을 1만큼 증가
  for i in range(0, M):
    for j in range(0, N):
      array[channel[i,j]//L]+=1
  
  #히스토그램으로 나타내기
  x = [x for x in range(L)]
  plt.plot(x, array, color=color)
  plt.xticks(range(0, 16, 1))
  plt.title("Histogram")
  plt.show()

def normal_histogram_16(channel, color):
  
  #배열 초기화
  array=[0 for i in range(L)]
  
  #히스토그램 칸을 1만큼 증가
  for i in range(0, M):
    for j in range(0, N):
      array[channel[i,j]//L]+=1
      
  #정규화
  for i in range(L):
    array[i] = array[i]/(M*N)
  
  #히스토그램으로 나타내기
  x = [x for x in range(L)]
  plt.plot(x, array, color=color)
  plt.xticks(range(0, 16, 1))
  plt.title("Normalized Histogram")
  plt.show()

histogram_16(r_img, "red")
histogram_16(g_img, "green")
histogram_16(b_img, "blue")

normal_histogram_16(r_img, "red")
normal_histogram_16(g_img, "green")
normal_histogram_16(b_img, "blue")
```
